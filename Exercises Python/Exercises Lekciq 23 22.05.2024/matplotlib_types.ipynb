```python
from matplotlib import pyplot as plt 
import numpy as np
```

Ploting various plot types with Matplotlib

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22,23,24,22,20,19,21]

plt.plot (days, temperatures, marker='o', linestyle = '-', color = 'b')
plt.title('Weekly Temperature Variation')
plt.xlabel('Days of the Week')
plt.ylabel("tEMPERATURE (°C)")
plt.grid(True)
plt.show()
```

Хистрограма - извежда зададените цифри ( в наши случай), разделя интервала от 60 до 90, а чрез bins се разделя на две. Във втория бин излизат 

```python
math_scores = [85,70,92,78,88,95,60,75,85,70]
print(f'min:{min(math_scores)}')
print(f'max:{min(math_scores)}')
print(f'len{min(math_scores)}')
print(f'sorted: {sorted(math_scores)}')
plt.hist(math_scores, bins=2)
```

Хистограма

```python
scores = np.array ([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6], dtype = 'int')

print(f'min:{min(scores)}')
print(f'max:{min(scores)}')
print(f'len:{min(scores)}')
print(np.histogram(scores, bins=4))

plt.hist(scores, bins=4, stacked = True)

```

Хистограма - БАР ЧАРТ 

```python
scores = np.array ([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6], dtype = 'int')

plt.hist(scores, bins = 4, edgecolor = 'black')

plt.xlabel('Test Scores')
plt.ylabel('Number of Students')
plt.ylabel('Science Scores')
plt.legend()
plt.grid(True)
plt.show()
```

Scatter plot 

```python
math_scores = [85, 70, 92, 78, 88, 95, 60, 75, 85, 70]
science_scores = [75, 65, 88, 72, 80, 92, 55, 70, 78, 68]

plt.scatter(math_scores, science_scores, c='blue', marker='o', label='Students')
plt.title('Math vs Science Test Scores')
plt.xlabel('Math Scores')
plt.ylabel('Science Scores')
plt.legend()
plt.grid(True)
plt.show()
```

BOX PLOT

```python
products = ['toys', 'food', 'drinks']
sales = [100,50,200]

plt.bar(products, sales, color = ['red', 'green', 'blue'], edgecolor = 'black')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Products Sales')
```

BOX PLOT - няма хистрограма, а хоризонтална кутийка, която се нарича така.

```python
scores = np.array ([2,2,2,2,2,3,3,4,5,5,5,6,6,6,6,6,6], dtype = 'int')

plt.boxplot(scores)

print(f'median:{np.median(scores)}')
print(f'quatile:{np.quantile(scores, [0.25, 0.5, 0.75, 1])}')
```

PIE CHART 

```python
budget_categories = ['Housing', 'Food', 'Transportation', 'Entertainment', 'Savings']
budget_alocation = [30, 20, 15, 10, 25]

plt.pie(budget_alocation, labels= budget_categories, autopct='%1.1f%%')
plt.title('MonthlyBudget Alocation')
```
