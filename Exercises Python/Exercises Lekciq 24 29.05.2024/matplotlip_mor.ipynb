```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create some data to be used 
x= np.linspace(1,100,10)
x
```

```python
# create 2 axes, sacked vertically:
# fig = plt.figure()
fig, [ax1, ax2] = plt.subplots(nrows=2, ncols=1, sharex=True, figsize = (10,6), dpi=100)

ax1= fig.add_subplot(211)
ax2 = fig.add_subplot(212)
# plot lines 
ax1.plot(x,x**2, color="red")
ax2.plot(x,x**3, color="green")
# set axes titles 
ax1.set_title("x**2")
ax1.set_title("x**3")

plt.tight_layout()
plt.savefig('figure_1280x720.jpg')
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z)

# Plot the line
ax.plot(x, y, z)

# Set labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Basic 3D Line Plot')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-2, 2, 5)
y = np.linspace(-2, 2, 5)
print(x)
print(y)
```

```python
xx, yy = np.meshgrid(x, y)
zz = xx + yy

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(xx, yy, zz, cmap="Accent")

# Add a color bar which maps values to colors
fig.colorbar(surface, shrink=0.5)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the Rosenbrock function
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

# Generate data
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title
ax.set_title('Surface Plot')

# Add a color bar which maps values to colors
fig.colorbar(surface, shrink=0.5)

# Show the plot
plt.show()
```
